```python
import json
```

```python
with open(r"A:\hypothesis_testingRAG\archive\hotpot_dev_distractor_v1.json", 'r', encoding="utf-8") as f:
    data = json.load(f)
print(len(data))
```

```python
sample = data[0]
print(sample.keys())
```

```python
sample['context']
```

```python
documents = []

for sample in data:
    for title, sentences in sample["context"]:

        text = " ".join(sentences)

        documents.append({
            "title": title,
            "text": text
        })

print(len(documents))
```

```python
documents[0]
```

```python
unique_docs = {}

for doc in documents:
    unique_docs[doc["title"]] = doc["text"]

documents = [
    {
        "title": k,
        "text": v
    }
    for k, v in unique_docs.items()
]

print(len(documents))
```

```python
import pandas as pd

df = pd.DataFrame(documents)

df.to_csv(
    "corpus.csv",
    index=False
)
```
